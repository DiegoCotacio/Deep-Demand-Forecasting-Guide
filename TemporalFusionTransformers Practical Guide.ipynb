#### Aquí nos detendremos para entender la data que un TFT puede procesar:

Existen 3 tipos de datos que son posibles incorporar:

1. **Series temporales**: 
Estos datos corresponden a los datos **historicos** base de una serie temporal: nuestro *timestamp* y *nuestra target*. Dado que podemos realizar agregaciones, incluiremos en esta lista el *SKU* y la *Sede*. Con esta información el modelo puede funcionar trabajando sobre los valores de target. Cuales son en nuestro dataset:

- date, volumen. (sku y agency)

2. **Covariables estaticas**:
Podemos considerar estos datos como "Metadata" o información que no varia en el tiempo pero que otorga un contexto al comportamiento de nuestra target. Cuales son:

- avg_population_2017, avg_yearly_household_income_2017   (podria ser also categoria, color, tamaño del sku)

3. **Covariables dinamicas**:
Esta es mi feature favorita. Estas son series de tiempo relacionadas al fenomeno de estudio, en este caso, el comportamiento del volumen de ventas. Pueden ser explicativas o causales al dar un contexto temporal a nuestra target. Aquí es donde puedes explorar con datos de eventos que suelan impactar tu problema de demanda.

  *Ojo*: estas pueden ser historicas o futuras. Cómo es esto? Digamos que hay eventos que sucedieron en el pasado que no podemos conocer en el futuro, pero otros si, como por ejemplo los días festivos. Llamemoslas series temporales relacionadas known y unknow. Esto sera clave al configurar el modelo.
Para nuesto caso son:

  **Unknow**: volume, industry_volume, soda_volume, avg_max_temp.

  **known**: price_regular, discount y los dias festivos basicamente: easter_day,good_friday,new_year,christmas,labor_day,independence_day,revolution_day_memorial, regional_games, fifa_u_17_world_cup, football_gold_cup, beer_capital, music_fest   

#### Ahora entendamos el scope del pronostico:

```python
# Una funcion simple para ver cuantos SKUs y Agencias existen

def count_classes_data(data, columns):
    class_counts = {}
    for column in columns:
        unique_classes = np.unique(data[column])
        class_counts[column] = len(unique_classes)
    return class_counts

selected_columns = ['sku', 'agency']
result = count_classes_data(data, selected_columns)
for var, count in result.items():
    print(f'{var}: {count} clases')
```

Resumen:
- Tenemos 25 SKU y 58 Agencias, es decir, 1450 pronosticos por realizar!
- Queremos predecir los siguientes 6 meses, estos es, 8700 puntos de datos.

*Nota: este proceso se puede desagregar al niveles mas granulares(semanas, dias, segundos). Asi por ejemplo, si queremos pronosticar la demanda de 120 SKU en 3 ciudades de forma semanal (7 dias) 4 veces al mes, esto seria: 1140 pronosticos (10 K puntos) mensuales*
    

#### Hagamos un poco de feature engineering para obtener más información que ayude al modelo a entender patrones

Vamos a agregar algunas features, modificar otras.

1. *time_idx*:  Incorporamos un indice de tiempo para cada mes dado que es nuestro periodo de referencia. El dataset contendra 59 meses.


2. *Otras*: month, log_volume, avg_volume_by_sku, avg_volume_by_agency


3. *Codificar los dias especiales*: Pasaremos de 0 y 1 a poner directamente el nombre del dia especial y si no lo es un "-". Dado que el modelo trabaja con Embeddings sera posible manejar este tipo de datos.

*nota: puedes continuar al siguiente punto, no es necesario que entiendas este proceso*

```python
# add time index
data["time_idx"] = data["date"].dt.year * 12 + data["date"].dt.month
data["time_idx"] -= data["time_idx"].min()

# add additional features
data["month"] = data.date.dt.month.astype(str).astype("category")  # categories have be strings
data["log_volume"] = np.log(data.volume + 1e-8)
data["avg_volume_by_sku"] = data.groupby(["time_idx", "sku"], observed=True).volume.transform("mean")
data["avg_volume_by_agency"] = data.groupby(["time_idx", "agency"], observed=True).volume.transform("mean")

# we want to encode special days as one variable and thus need to first reverse one-hot encoding
special_days = [
    "easter_day",
    "good_friday",
    "new_year",
    "christmas",
    "labor_day",
    "independence_day",
    "revolution_day_memorial",
    "regional_games",
    "fifa_u_17_world_cup",
    "football_gold_cup",
    "beer_capital",
    "music_fest",
]
data[special_days] = data[special_days].apply(lambda x: x.map({0: "-", 1: x.name})).astype("category")
```

## 2. Preparación de datos para entrenamiento del TFT

#### Crearemos nuestro Pytorch Dataset  a partir del dataframe y tambien los Dataloaders


Que son y por qué utilizarlos:

* **TimeSeriesDataset**:

esta es una clase para estructurar los datos de forma que sean compatibles con el modelo. Lo usaremos para definir cuales son nuestras variables categoricas/numericas estaticas o dinamicas (know y unknow). Asimismo especificamos el metodo para normalizar los datos (GroupNormalizer)

* **Dataloaders**:
Basicamente es un objeto que maneja de forma mas eficiente los datos en pequeños lotes durante el entrenamiento y asi evitar cargar todo el conjunto de datos en la memoria al mismo tiempo. Tendremos dataloader de train y de validación.

```python
# Definimos la longitud de nuestros datasets de train y val

max_prediction_length = 6   #cuanto hacia adelante queremos predecir en val (6 meses)
max_encoder_length = 24     # que tan hacia atras queremos ver (24 meses)
training_cutoff = data["time_idx"].max() - max_prediction_length #punto de corte entre de val, el resto atras es para train
```

#### Crearemos nuestro pytorch dataset para training

```python
# Ahora instanciamos el objeto training que contendra las especificacion de nuestro dataset de entrenamiento

training = TimeSeriesDataSet(
    data[lambda x: x.time_idx <= training_cutoff],    #datos de entrenamiento del dataframe procesado
    time_idx="time_idx",                              # el time_idx 
    target="volume",                                  #nuestra target a predecir
    group_ids=["agency", "sku"],                      #nuestos niveles de agregación
    min_encoder_length=max_encoder_length // 2,  # keep encoder length long (as it is in the validation set)
    max_encoder_length=max_encoder_length,            #cuanto atras vamos a ver (24)
    min_prediction_length=1,                          #esto es cuanto minimo vamos a ver hacia a tras
    max_prediction_length=max_prediction_length,      #cuantos meses vamos a predecir (6)
    static_categoricals=["agency", "sku"],            # nuestras var estaticas categoricas
    static_reals=["avg_population_2017", "avg_yearly_household_income_2017"], #var estaticas reales
    time_varying_known_categoricals=["special_days", "month"], #var dinamicas know
    variable_groups={"special_days": special_days},  # groupo de var categoricas se tratan como una variable
    time_varying_known_reals=["time_idx", "price_regular", "discount_in_percent"], #var reales dinamicas know
    time_varying_unknown_categoricals=[],  #categoricas unknow
    time_varying_unknown_reals=[    
        "volume",
        "log_volume",
        "industry_volume",
        "soda_volume",
        "avg_max_temp",
        "avg_volume_by_agency",
        "avg_volume_by_sku",
    ],                                               #numericas unknow
    target_normalizer=GroupNormalizer(
        groups=["agency", "sku"], transformation="softplus"  
    ),                                               #tipo de normalizacion con funcion de activacion softplus
    add_relative_time_idx=True, 
    add_target_scales=True,
    add_encoder_length=True,
) # estas tres sirven para controlar si se agregan features adicionales relacionadas con time_dix, escalas y long de enc

```

#### Ahora nuestro pytorch dataset para validation

```python
#dado que el objeto training ya contiene la config del train dataset, creamos validation dataset a partir del mismo
validation = TimeSeriesDataSet.from_dataset(training, data, predict=True, stop_randomization=True)
```

#### Y ahora nuestros dataloaders para train y val

```python
#dos parametros claves:
# batch_size: numero de muestras procesadas simultaneamente, es el lo mismo que el tamaño de los mini lotes
# num_workers: numero de subprocesos utilizados para cargar los datos en paralelo

batch_size = 128  # set this between 32 to 128
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers=0)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size * 10, num_workers=0)
```

#### Listo, antes de entrenar un TFT, creemos un baseline

Realizaremos predicciones para los proximos 6 meses repitiendo el ultimo volumen observado. Comparemos las metricas mas adeltante con las del TFT.

```python
# calculate baseline mean absolute error, i.e. predict next value as the last available value from the history
baseline_predictions = Baseline().predict(val_dataloader, return_y=True)
```

```python
from pytorch_forecasting.metrics import RMSE, SMAPE, MAPE, QuantileLoss, PoissonLoss

mae = MAE()(baseline_predictions.output, baseline_predictions.y)
smape = SMAPE()(baseline_predictions.output, baseline_predictions.y)
rmse = RMSE()(baseline_predictions.output, baseline_predictions.y)
#quantile_loss = QuantileLoss()(baseline_predictions.output, baseline_predictions.y)
poisson_loss = PoissonLoss()(baseline_predictions.output, baseline_predictions.y)

print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"SMAPE: {smape:.4f}")
#print(f"Quantile Loss: {quantile_loss:.4f}")
print(f"Poisson Loss: {poisson_loss:.4f}")
```

## 3. Entrenamiento de nuestro Temporal Fusion Transformer

Hasta ahora ya hemos explicado el tipo de datos que un TFT puede procesar. Ahora explicaremos la arquitectura y entenderas mejor como este modelo maneja esta tipologia:

#### Explicación de Alto Nivel:

La arquitecutra TFT esta diseñada para procesar de forma independiente los 3 tipos de datos ya mencionados: *los datos de la serie temporal base (es decir, el comportamiento pasado), covariables estaticas (metadata o info que no varia en el tiempo) y series relacionadas que pueden ser know y unknow, algunas de ellas hacia el futuro*. Estos son los 3 componentes temporales: *features del pasado, del futuro y estaticas*. Puedes notarlo en la siguiente grafica.


Al iniciar la creacion de un predictor para foracasting (un TFT entrenado), se ejecutan multiples operaciones internas, de forma resumida y secuencial estas serian:

1. Entiende el peso/importancia que tiene cada feature en dataset y silencia las mas debiles, esto permite maximizar 

2. Realiza un encoding de las features estaticas pasandolas por los diferentes momentos enriqueciendo el proceso.

3. Busca patrones de corto plazo empleando redes LSTM (por ejemplo patrones diarios o semanales recientes)

4. Y para el largo plazo emplea un mecanismo de atención sobre distintos pasos del tiempo que sean relevantes.

5. Por ultimo la generación de forecast es intercuantilica, lo que nos facilita obtener intervalos de confianza que van entre: 0.1, 0.25, 0.5, 0.75 y 0.9

#### Configuración general de nuestro objeto trainer

```python
import tensorflow as tf
```

```python
# early_stop_callback: este es el callback para detener condiciones cuando se cumplan las siguientes condiciones:
# val_loss es la metricas que monitorizamos
# min_delta es un umbral minimo. Si la diferencia entre metrica actual y mejor registrada no lo supera asume que no hay mejor
# patience indica cuantas epocas se espera despues de que la metrica monitorizda no mejore antes de detener training
# mode es para indicar si buscamos minimizar o maximizar la metrica val loss

early_stop_callback = EarlyStopping(monitor="val_loss", min_delta=1e-4, patience=10, verbose=False, mode="min")

lr_logger = LearningRateMonitor()  # logeamos el learning rate

logger = TensorBoardLogger("lightning_logs")  # logeamos en Tensorboard (podras ver resultados en este folder)
```

#### Ahora crearemos nuestro objeto trainer el cual sera el parametro clave del modelo

Cuales son sus parametros?

* **max_epochs**: numero maximo de epocas de entrenamiento (clave)
* **accelerator**: el dispositivo de aceleracion, puede ser cpu o gpu
* **enable_model_summary**: es para imprimir el resumen del modelo (parametros, capas, etc)
* **gradient_clip_val**: valor maximo para la normal del gradiente, estabiliza el entrenamiento
* **limit_train_batches**: limite de numero de lotes de entrenamiento por epoch.
* **callbacks**: la lista de devoluciones fijadas anteriormente para logging y parada temprana
* **logger**: registra la info de entrenamiento en TensorBoard

```python
trainer = pl.Trainer(
    max_epochs=50,
    accelerator="cpu",
    enable_model_summary=True,
    gradient_clip_val=0.1,
    limit_train_batches=50,  # coment in for training, running valiation every 30 batches
    # fast_dev_run=True,  # comment in to check that networkor dataset has no serious bugs
    callbacks=[lr_logger, early_stop_callback],
    logger=logger,
)
```

#### Ahora creamos nuestro TFT

Veamos los parametros:

* **training**: nuestro objeto pytorch dataset para training
* **learning_rate**: es el learning rate 
* **hidden_size**:  es el tamaño de la capa oculta en el modelo
* **attention_head_size**: Es el tamaño de las cabezas de atención en la capa de atención
* **dropout**:  proporcion de dropout aplicada en el modelo para evitar overfitting
* **hidden_continuous_size**: tamaño de capa oculta para variables continuas
* **loss**: nuestra funcion de perdida 
* **log_interval**:  muestra cada cuantos lotes se muestra info registrada
* **optimizer**: es un optimizador para entrenar el modelo.
* **reduce_on_plateau_patience**: la espera para reducir learning rate cuando la metricas de validacion deje de mejorar

```python
tft = TemporalFusionTransformer.from_dataset(
    training,
    learning_rate=0.03,
    hidden_size=16,
    attention_head_size=2,
    dropout=0.1,
    hidden_continuous_size=8,
    loss=QuantileLoss(),
    log_interval=10,  # uncomment for learning rate finder and otherwise, e.g. to 10 for logging every 10 batches
    optimizer="Ranger",
    reduce_on_plateau_patience=4,
)
print(f"Number of parameters in network: {tft.size()/1e3:.1f}k")
```

#### Entrenemos!!

Ahora le pasamos al objeto trainer la config del modelo TFT y los dataloaders de train y val

```python
trainer.fit(
    tft,
    train_dataloaders=train_dataloader,
    val_dataloaders=val_dataloader,
)
```

#### Nuestros peores candidatos:

### 1. Configurar el dataset de Salida (Encoder + Decoder = new_prediction_data ):



* **Dos conceptos claves: Encoder y Decoder**: \
  El **encoder** es esencialmente el dataframe preprocesado (el que quedó arriba llamado "data") y el **decoder** es un dataframe creado a partir del encoder para almacenar las predicciones (*a partir de la ultima fecha o indice del encoder obviamente*). La fusión de los dos es **new_prediction_data**. Este ultimo sera el que pasaremos al modelo para crear los pronosticos.

  Esto que significa? Que para poder realizar predicciones necesito conocer las mismas variables que usé en el entrenamiento del TFT.

Ahora la operación más importante y clave de entender.
Haremos una operaciones de crear un dataframe usando List Comprehensions:


* **Que haremos?**:
Crea un dataframe llamado "decoder_data" que es el resultado de concatenar varios versiones modificadas del dataframe "last_data". Cada version modificada avanza una cierta cantidad de meses (i) en la columna de fechas (date) y se realiza en un rango hasta 6. El resultado final es el dataframe decoder.


* *pd_concat*: concatenaremos una lista de dataframes.


* *[...]*: Creamos una lista de dataframes en funcion de la variable i, la cual va de 1 hasta max_prediction_length (6).


* *last_data.assign(date=lambda x: x.date + pd.offsets.MonthBegin(i))*: Con el metodo assign creamos copias modificadas del dataframe last_data. La modificacion consiste en sumar i meses al valor de la columna 'date'. pd.offsets.MonthBegin(i) indica que se debe avanzar i meses desde la fecha actual, asegurandose de que las fechas resultantes sean el primer dia del mes.


* *for i in range(1, max_prediction_length + 1)*: Es un bucle que recorre valores de 1 a 6 (los pasos de prediccion)



* *ignore_index=True*: Se reindexa el dataframe resultante para que los indices sean secuenciales desde cero en lugar de mantener los indices originales

```python
# Ahora si iniciemos el decoder:
decoder_data = pd.concat(
    [last_data.assign(date=lambda x: x.date + pd.offsets.MonthBegin(i)) for i in range(1, max_prediction_length + 1)],
    ignore_index=True,
)
```

Ahora crearemos un indice  de tiempo para que coincida con el indice de tiempo del decoder.

Primero extraemos y multiplicamos el año del encoder x 12 y le sumamos el mes para de esta forma desplazarnos dentro del año.

Despues ajustamos los valores del indice de tiempo calculando un "desplazamiento que nos permita hacer esto. Para esto obtenemos el valor maximo de time_idx del encoder y el valor minimo de time_idx del decoder, una vez esto realizamos una resta para calcular cuantos pasos de tiempo deben ser ajustados para que los indices de decoder sean coherentes con los de encoder. (OJO: sumamos 1 para asegurarnos de que el ajuste sea correcto).

```python
# add time index consistent with "data"
decoder_data["time_idx"] = decoder_data["date"].dt.year * 12 + decoder_data["date"].dt.month
decoder_data["time_idx"] += encoder_data["time_idx"].max() + 1 - decoder_data["time_idx"].min()
```

#### Realizaremos las siguientes operaciones:


1. Crear 3 columnas para almacenar la info que extraeremos (la llenaremos mientras con nan)


2. Extraeremos las posiciones de los quantiles con numpy (extraremos 3,4 y 5 que corresponden a 0.25,0.5 y 0.75)


4. Para cada columna usaremos un bucle for para ingestar los valores de cada array

```python
#pasos 1.
predictions_df["q50_volume_prediction"]=np.nan
predictions_df["q25_volume_prediction"]=np.nan
predictions_df["q75_volume_prediction"]=np.nan

# Paso 2: Obtener las predicciones para los cuantiles específicos
q50_predictions = new_prediction_results.output.prediction.cpu().numpy()[:,:,4]
q25_predictions = new_prediction_results.output.prediction.cpu().numpy()[:,:,3]
q75_predictions = new_prediction_results.output.prediction.cpu().numpy()[:,:,5]

# Paso 3: Iterar a través de los resultados del pronóstico y agregar los valores a las columnas correspondientes
for i, row in new_prediction_results.index.iterrows():
    predictions_df.loc[
        (predictions_df["agency"]==row["agency"]) & 
        (predictions_df["sku"]==row["sku"]), 
        "q50_volume_prediction"] = q50_predictions[i]
    
    predictions_df.loc[
        (predictions_df["agency"]==row["agency"]) & 
        (predictions_df["sku"]==row["sku"]), 
        "q25_volume_prediction"] = q25_predictions[i]
    
    predictions_df.loc[
        (predictions_df["agency"]==row["agency"]) & 
        (predictions_df["sku"]==row["sku"]), 
        "q75_volume_prediction"] = q75_predictions[i]
```

```python
# el ouput es un float que podemos redondear:

float_columns = predictions_df.select_dtypes(include=['float64']).columns
predictions_df[float_columns] = predictions_df[float_columns].round(0)
```
